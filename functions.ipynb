```python
def do_something(n):
  total=0
  i=1
  while i<=n:
    total += i
    i += 1
  return total

print(do_something(5))
```

```python
def gcd(a,b):
  x=min(a,b)
  for i in range(x,0,-1):
    if a%i==0 and b%i==0:
      print(i)
      break
```

```python
gcd(5,15)
```
